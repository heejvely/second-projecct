```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/2차 의도분류/kobert 실습/sample_data_3000.csv')
data.head()
```

```python
data[data.comment.str.contains('\n')]
```

```python
data['comment'] = data.comment.replace('\n',' ',regex=True)
```

```python
neg_train = data[:800]
neg_valid = data[800:900]
neg_test = data[900:1000]

neu_train = data[1000:1800]
neu_valid = data[1800:1900]
neu_test = data[1900:2000]

pos_train = data[2000:2800]
pos_valid = data[2800:2900]
pos_test = data[2900:3000]
```

```python
train = pd.concat([neg_train, neu_train,pos_train],ignore_index=True)
valid = pd.concat([neg_valid, neu_valid,pos_valid],ignore_index=True)
test = pd.concat([neg_test, neu_test,pos_test],ignore_index=True)
```

```python
import numpy as np
np.random.permutation(train)
```

```python
import sklearn

sklearn.utils.shuffle(train).to_csv('/content/drive/MyDrive/2차 의도분류/kobert 실습/train_3000.txt',sep='\t',header=False,index=False)
sklearn.utils.shuffle(valid).to_csv('/content/drive/MyDrive/2차 의도분류/kobert 실습/valid_3000.txt',sep='\t',header=False,index=False)
sklearn.utils.shuffle(test).to_csv('/content/drive/MyDrive/2차 의도분류/kobert 실습/test_3000.txt',sep='\t',header=False,index=False)
```
