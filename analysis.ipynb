```python
df1 = df.dropna(axis=0)
df1
```

```python
real_cat = pd.pivot_table(df1, index='cat', columns = 'score',values='comment', aggfunc='count').reset_index()
```

```python
pred_cat = pd.pivot_table(df1, index='cat', columns = 'emotion',values='comment', aggfunc='count').reset_index()
```

```python
pred_cat
```

```python
real_cat['sum'] = real_cat[0] +real_cat[1] + real_cat[2]
pred_cat['sum'] = pred_cat[0] +pred_cat[1] + pred_cat[2]
```

```python
for i in [0,1,2]:
  real_cat[f'{i}_rate'] = real_cat[i]/real_cat['sum']

real_cat
```

```python
for i in [0,1,2]:
  pred_cat[f'{i}_rate'] = pred_cat[i]/pred_cat['sum']

pred_cat
```

```python
real_cat_jung = pd.pivot_table(df1, index='cat_jung', columns = 'score',values='comment', aggfunc='count').reset_index()
pred_cat_jung = pd.pivot_table(df1, index='cat_jung', columns = 'emotion',values='comment', aggfunc='count').reset_index()

real_cat_jung['sum'] = real_cat_jung[0] +real_cat_jung[1] + real_cat_jung[2]
pred_cat_jung['sum'] = pred_cat_jung[0] +pred_cat_jung[1] + pred_cat_jung[2]

for i in [0,1,2]:
  real_cat_jung[f'{i}_rate'] = real_cat_jung[i]/real_cat_jung['sum']

for i in [0,1,2]:
  pred_cat_jung[f'{i}_rate'] = pred_cat_jung[i]/pred_cat_jung['sum']
```
